```python
from rdkit import Chem
from rdkit.Chem.Draw import IPythonConsole
from murcko_hist import murcko_hist
IPythonConsole.molSize = (400, 300)
IPythonConsole.drawOptions.addAtomIndices = True
# IPythonConsole.drawOptions.addBondIndices = True
%load_ext autoreload
%autoreload 2
```

```python
mol = Chem.MolFromSmiles('O=C(O)[C@@H]1/N=C(\SC1)c2sc3cc(O)ccc3n2')
murcko_hist(mol, show_mol_scaffold=True)
```

```python
mol = Chem.MolFromSmiles('CC1=CC(C2CC(=O)OC2=O)CC2C(=O)OC(=O)C12')
murcko_hist(mol, show_mol_scaffold=True)
```
