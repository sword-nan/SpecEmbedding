```python
import numpy as np

from matchms.importing import load_from_msp
from matchms.filtering import select_by_mz, normalize_intensities 
from utils import default_filters

def filter(s):
    s = default_filters(s)
    s = normalize_intensities(s)
    return s

query_spectra = list(load_from_msp("/data1/xp/data/MSBert/MTBLS1572/q.msp"))
ref_spectra = list(load_from_msp("/data1/xp/data/MSBert/MTBLS1572/r.msp"))
query_spectra = [filter(s) for s in query_spectra]
ref_spectra = [filter(s) for s in ref_spectra]

np.save("/data1/xp/data/MSBert/MTBLS1572/query.npy", query_spectra)
np.save("/data1/xp/data/MSBert/MTBLS1572/ref.npy", ref_spectra)
```
